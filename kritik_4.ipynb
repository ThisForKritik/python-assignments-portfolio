```python
import numpy as np
import matplotlib.pyplot as plt

# The function
def f(x):
    y = x**2
    return y

# The derivative
def d(x):
    m = 2*x
    return m

# Parameters for base function
n = 100
a = -10
b = 10
x_base = np.linspace(a, b, n+1)

# Parameters for gradient
r = 0.8
points = 20 # The number of gradient points to generate
initial_point = -10
x_coords = [initial_point]
y_coords = [f(initial_point)]
m_coords = [d(initial_point)]

# Create the gradient
for i in range(0, points):
    x = x_coords[i] - r*m_coords[i]
    y = f(x)
    m = d(x)
    x_coords.append(x)
    y_coords.append(y)
    m_coords.append(m)
    
print("minima at (", round(x_coords[points]), ",", round(y_coords[points]), ")")
print("initial point at (", x_coords[0], ",", y_coords[0], ")")
print("learning rate =", r)

# Plot base function and gradient
plt.plot(x_base, f(x_base))
plt.plot(x_coords, y_coords)
plt.grid()
plt.show()
```
