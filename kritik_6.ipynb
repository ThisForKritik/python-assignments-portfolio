```python
import numpy as np
import matplotlib.pyplot as plt

y_values = []
def normal_density(mean, variance, x):
    y = (1/np.sqrt(2*np.pi*variance**2)) * np.exp(-(x-mean)**2 / (2*variance**2))  
    y_values.append(y)

x = 1
normal_density(1, 10, x) # condition 1
normal_density(10, 1, x) # condition 2
normal_density(1, 1, x) # condition 3

plt.scatter(x,y_values[0], label="condition 1")
plt.scatter(x,y_values[1], label="condition 2")
plt.scatter(x,y_values[2], label="condition 3")
plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.axis("equal")
plt.grid()
plt.show()
```
